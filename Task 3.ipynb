# DATA VISUALIZATION

```python
print("Find correlation between different numerical variables")
Sample.corr()
```

```python
plt.figure(figsize=(12,4))
sns.heatmap(Sample.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(12,6))
plt.subplot(221); sns.countplot('Category',data=Sample)        
plt.subplot(222); sns.barplot(x='Category',y='Profit',data=Sample)
plt.subplot(223); sns.countplot('Ship Mode',data=Sample)
plt.subplot(224); sns.barplot(x='Ship Mode',y='Profit',data=Sample)
```

```python
plt.figure(figsize=(10,5))
Sample['Region'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data= Sample)
plt.title('Quantity v/s Profit')
plt.legend()
plt.show
```

```python
sns.lineplot(x='Discount',y='Profit',data=Sample)
plt.title('Discount v/s Profit')
plt.show()
```

```python
sns.lineplot(x='Sales',y='Profit',data=Sample)
plt.title('Sales v/s Profit')
plt.show()
```

```python
fig, axs = plt.subplots(ncols=2, nrows = 2, figsize = (10,10))
sns.distplot(Sample['Sales'], color = 'red',  ax = axs[0][0])
sns.distplot(Sample['Profit'], color = 'green',  ax = axs[0][1])
sns.distplot(Sample['Quantity'], color = 'orange',  ax = axs[1][0])
sns.distplot(Sample['Discount'], color = 'blue',  ax = axs[1][1])
axs[0][0].set_title('Sales Distribution', fontsize = 20)
axs[0][1].set_title('Profit Distribution', fontsize = 20)
axs[1][0].set_title('Quantity distribution', fontsize = 20)
axs[1][1].set_title('Discount Distribution', fontsize = 20)
plt.show()
```

# Citywise Analysis of the Profit


```python
Sample_city= Sample.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean()
Sample_city = Sample_city.sort_values('Profit')
Sample_city.head()
```

```python
#1.Low Profit
Sample_city['Profit'].head(30).plot(kind='bar',figsize=(15,5),color = 'blue')
plt.title('City wise analysis of Sale, Discount, profit')
```

```python
#2. High Profit
Sample_city['Profit'].tail(30).plot(kind='bar',figsize=(15,5),color = 'green')
plt.title('City wise analysis of Sale, Discount, profit')
```
